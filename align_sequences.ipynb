```python
import os
import subprocess
import tempfile

def perform_alignment(input_file, output_file):
    # Define the command for ClustalW
    command = f"clustalw -infile={input_file} -outfile={output_file}"

    # Execute ClustalW
    subprocess.run(command, shell=True, check=True)

if __name__ == "__main__":
    # Specify the file containing database sequences
    database_file = "dog_breeds.fa"

    # Create a temporary directory to store intermediate files
    temp_dir = tempfile.mkdtemp()

    # Define the input and output file paths
    input_file = database_file
    output_file = os.path.join(temp_dir, "aligned_sequences.aln")

    # Perform sequence alignment
    perform_alignment(input_file, output_file)

    print("Alignment completed. Output file:", output_file)
```
